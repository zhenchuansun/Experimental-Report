```python
import matplotlib.pyplot as plt
import numpy as np

# 设置 matplotlib 支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 实验数据
L = np.array([644, 527, 727])
I_exp = np.array([40.082, 40.2456, 39.9408])
I_theory = 39.591159

# 计算误差（实验值与理论值的差）
error = I_exp - I_theory

# 绘制实验转动惯量与理论转动惯量对比图
plt.figure(figsize=(8,6))
plt.plot(L, I_exp, 'o', label='实验转动惯量')
plt.axhline(I_theory, color='red', linestyle='--', label='理论转动惯量')
# 在数据点旁标注误差值
for x, y, e in zip(L, I_exp, error):
    plt.text(x, y, f'{e:+.3f}', fontsize=10, ha='center', va='bottom')

plt.xlabel("摆线长度 (mm)", fontsize=12)
plt.ylabel("转动惯量 (×10-5 kg·m2)", fontsize=12)
plt.title("不同摆线长度下转动惯量实验值与理论值比较", fontsize=14)
plt.legend(loc='lower right', fontsize=14)
plt.xlim(500, 800)  # 设置x轴范围
plt.ylim(30, 42)  # 设置y轴范围
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

# 标记理论值
plt.text(800, I_theory, f'理论值: {I_theory:.3f}', fontsize=14, color='red', ha='right', va='bottom')
plt.show()

# 绘制误差分布图
plt.figure(figsize=(8,6))
plt.plot(L, error, 'o')
plt.xlabel("摆线长度 (mm)")
plt.ylabel("误差 (×10-5 kg·m2)")
plt.title("不同摆线长度下误差分布")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 0.8)  # 设置y轴范围
plt.grid(True)
plt.show()

# 输出误差统计信息
print("各数据点误差 (实验值 - 理论值)：", error)
print("误差均值：", np.mean(error))
print("误差标准差：", np.std(error))
```
