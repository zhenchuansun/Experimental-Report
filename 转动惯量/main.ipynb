```python
import matplotlib.pyplot as plt
import numpy as np

#  matplotlib 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 
L = np.array([644, 527, 727])
I_exp = np.array([40.082, 40.2456, 39.9408])
I_theory = 39.591159

# 
error = I_exp - I_theory

# 
plt.figure(figsize=(8,6))
plt.plot(L, I_exp, 'o', label='')
plt.axhline(I_theory, color='red', linestyle='--', label='')
# 
for x, y, e in zip(L, I_exp, error):
    plt.text(x, y, f'{e:+.3f}', fontsize=10, ha='center', va='bottom')

plt.xlabel(" (mm)", fontsize=12)
plt.ylabel(" (10-5 kgm2)", fontsize=12)
plt.title("", fontsize=14)
plt.legend(loc='lower right', fontsize=14)
plt.xlim(500, 800)  # x
plt.ylim(30, 42)  # y
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

# 
plt.text(800, I_theory, f': {I_theory:.3f}', fontsize=14, color='red', ha='right', va='bottom')
plt.show()

# 
plt.figure(figsize=(8,6))
plt.plot(L, error, 'o')
plt.xlabel(" (mm)")
plt.ylabel(" (10-5 kgm2)")
plt.title("")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 0.8)  # y
plt.grid(True)
plt.show()

# 
print(" ( - )", error)
print("", np.mean(error))
print("", np.std(error))
```
